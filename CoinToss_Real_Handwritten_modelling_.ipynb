#Explanation
The visual inspection suggests that the Sequence1 is more likely to be human generated because it has long sequences of H's and T's. This means that the previous event provides a suggestion as to what the next event would be which is a matric for the evaluate() function. The evaluate() decides if the sequence is real or handwritten. 

Assumption: Sequences with longer subsequences of H/T in it are Handwritten.



### #1
So, I started with the first event of the sequence and then grouped the sequence such that all the subsequent H's after a H form one group. The appearance of T would end the H group and forms a new T group. It is illustrated with an example below:

[H,H,H,T,T,H,T,T] is grouped as [HHH,TT,H,TT]

This section is handled by the user-defined cluster() function.

# #2

Next, keeping in mind that the handwritten sequence has longer subsequences of H's and T's, we punish such sequences. As a result, the sequence which has a higher score generated by the score() function will be regarded as handwritten. 

The score is given by squaring up [(length of H's or T's)-1] so that longer subsequences of all H's or T's are hammered more. The intuition behind the subtraction of 1 is explained as follows:

Suppose the sequence is [H,H,H,T,T,T,H,T]. We start a random guess and it does not add any score to sequence because it is completely random. Next, we will check if the prediction is True. If it is true, the guess for the next element of the sequence is the same. And if it is false, the guess would be the opposite and the same process is followed. This approach is bound to make a mistake when it encounters the opposite flip. The mistake is adjusted by the subtraction of 1 in the next subsequence.

So the sequence assigns the value 2 for the 3 H's which accounts for 1 "degree of mistake", 2 for the 3 T's that follow as the (-1) account for the mistake it would make while predicting the first T, 0 for the 1 H and 0 for the 1 T for the same reason. Hence, the score is the sum of the squared of 2,2,0,0 which is 8. For a more random sequence like [H,H,T,H,T,T,H,T], the score would be 2.  


# #3

The evaluate function then decides which sequence is handwritten by comparing the score of each sequence. The sequence with greater score is handwritten.

```python
#getting an idea about the probalility distribution
print('Sequence 1 : count_H = ', seq1.count('H') , 'out of ', len(seq1)) 
print('Sequence 2 : count_H = ', seq2.count('H'), 'out of ', len(seq2))
```

```python
def cluster(seq): 
#this function forms subsequences of H's and T's only. 
    seqq=[]
    k=0
    str=""
    for i in range(len(seq)):
        if seq[k]==seq[i]:
            str=str+seq1[i]
        else:
            seqq.append(str)
            k=i
            str=""
            str=str+seq[i]
    seqq.append(str)
    return seqq


def score(sequence):
#score assigns sum of squares score to the sequence
  sum_of_square = 0 
  len_sq = 0
  for x in sequence:
    if (len(x)-1) != 0:
      len_sq = (len(x)-1)*(len(x)-1)
      sum_of_square = sum_of_square + len_sq
  return sum_of_square


def evaluate(sequence1,sequence2):
#evaluates which sequence is handwritten based on
  combined_seq1 = cluster(seq1)
  combined_seq2 = cluster(seq2)

  score1 = score(combined_seq1)
  score2 = score(combined_seq2)

  print('score of sequence1 =', score1)
  print('score of sequence2 =', score2)
  if score1 > score2:
    print('Sequence1 is handwritten')
  else:
    print('Sequence2 is handwritten')
```
